```python
import pandas as pd

park_data = pd.read_csv('bases/parkinsons.data', header=None)

park_data
```

```python
feature_names = park_data.iloc[0]
feature_names.pop(0)

df = park_data.drop(park_data.index[0])
df = df.drop(0, axis=1)
df = df.astype(float)

df
```

```python
feature_names
```

```python
df[17].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import StratifiedKFold

features = list(df.columns.values)
features.remove(17)

X = df.drop(columns=17, axis=1).values
y = df[17].values

std = StandardScaler()
X = std.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)

sm = SMOTE()
X_train,y_train = sm.fit_resample(X_train,y_train)

kfold = StratifiedKFold(n_splits= 10, random_state= 0, shuffle=True)
```

```python
import numpy as np

value_counts = np.unique(y_train, return_counts=True)

print(value_counts)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from pandas import DataFrame

model = SVC(kernel='linear')

model_results = []

for train_index, test_index in kfold.split(X_train, y_train):
    
    X_fold_train, y_fold_train = X_train[train_index], y_train[train_index]
    model.fit(X_fold_train, y_fold_train)
    y_pred = model.predict(X_fold_train)
    model_results.append(accuracy_score(y_fold_train, y_pred))

model_results = np.array(model_results)

print("SVM linear Accuracy (treino): :", model_results.mean() )
df_teste = pd.DataFrame(model_results)
df_teste
```

```python
from sklearn.model_selection import cross_val_score

model_test_results = cross_val_score(model, X_train, y_train, cv=kfold, n_jobs= 4, scoring="accuracy")

print("SVM linear Accuracy (teste): ", model_test_results.mean())
pd.DataFrame(model_test_results)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
cm = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))

print("Validação: ", accuracy_score(y_test, y_pred))
cm.plot()
```

```python
import shap

shap.initjs()

explainer = shap.KernelExplainer(model.predict, X)
shap_values = explainer.shap_values(X)
shap.summary_plot(shap_values, X_train, plot_type="bar")
```

```python
shap_sum = np.abs(shap_values).mean(axis=0)
importance_df = pd.DataFrame([features, shap_sum.tolist()]).T
importance_df.columns = ['feature', 'importance']
importance_df = importance_df.sort_values('importance', ascending=False)

print("Features: ", len(features))
print(importance_df)
```

```python
mean = importance_df['importance'].mean()
std = importance_df['importance'].std()

values = importance_df['importance']

z_scores = [(val - mean) / std for val in values]
```

```python
n = len([x for x in z_scores if x > 0])
ft = list(importance_df['importance'].head(n).index.astype(int))

X_train_shap = X_train[:, ft]
X_test_shap = X_test[:, ft]

print("Best Features: ", ft)
print(X_train_shap.shape)
```

```python
shap_results = cross_val_score(model, X_train_shap, y_train, cv=kfold, n_jobs= 4, scoring="accuracy")

print("SVM linear Accuracy (SHAP): ", shap_results.mean())
pd.DataFrame(shap_results)
```

```python
model.fit(X_train_shap, y_train)
y_pred = model.predict(X_test_shap)
cm = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))

print("Validação ", accuracy_score(y_test, y_pred))
cm.plot()
```

TESTE 3 - COM AUTOML

```python
from autosklearn.classification import AutoSklearnClassifier
from autosklearn.metrics import balanced_accuracy, precision, recall, f1
from sklearn.metrics import accuracy_score

minutos = 5

automl = AutoSklearnClassifier(
    include = {
        "classifier": ["random_forest", "extra_trees", "decision_tree", 
                       'liblinear_svc', 'libsvm_svc','k_nearest_neighbors']
    },
    time_left_for_this_task= minutos*60,
    per_run_time_limit= 30,
    scoring_functions= [balanced_accuracy, precision, recall, f1],
    ensemble_class= 'none',
    #ensemble_nbest = 25,
    n_jobs = 4
)

automl.fit(X_train, y_train, dataset_name= 'Parkinson Disease Prediction')
```

```python
def get_metric_result(cv_results):
    results = pd.DataFrame.from_dict(cv_results)
    results = results[results["status"] == "Success"]
    cols = ["rank_test_scores", "param_classifier:__choice__", "mean_test_score"]
    cols.extend([key for key in cv_results.keys() if key.startswith("metric_")])
    return results[cols]
```

```python
print("Metric results")
print(get_metric_result(automl.cv_results_).to_string(index=False))
```
